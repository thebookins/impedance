# Input impedance of simple pipes
This brief notebook contains some input impedance measurements for simple pipes (open and closed); some of the work behind [this paper](https://www.phys.unsw.edu.au/jw/reprints/DickensSmithWolfe.pdf).

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

## Closed pipe, 200 mm long
15 mm diameter PCV pipe, 200 mm long, closed at the far end. The input impedance of the pipe was measured at 25 degrees C, and a relative humidity of 0.43.

```python
df = pd.read_csv('data/closed_pipe_200mm.csv', index_col=0)
df['Z'] = df['Re(Z)'] + 1j * df['Im(Z)']
```

```python
df.head()
```

```python
fig, axes = plt.subplots(2, 1, sharex=True, figsize=(10, 10))

axes[0].semilogy(df.index, np.abs(df['Z']));
axes[0].set_ylabel('|Z| (Pa s m^-3)')

axes[1].plot(df.index, np.angle(df['Z']));
axes[1].set_xlabel('f (Hz)')
axes[1].set_ylabel('arg(Z)');
```
